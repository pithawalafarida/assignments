```python

#import the libraries
import pandas as pd
#from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt

#importoing the data.Data is taken from the site 'https://www.biz.uiowa.edu/faculty/jledolter/datamining/dataexercises.html'
data = pd.read_csv('weather.csv')
#printing the top 5 rows
data.head()
#looking at the features
data.columns
#the data types of features
data.dtypes

#notice that RISK_MM relation with the RainTomorrow variable are equivalent
rain_mm=data[data['RISK_MM']>1]
raintom=data[data['RainTomorrow'] == 'Yes']
raintom.equals(rain_mm)


#The Rainfall variable and the RainToday are equivalent
rain=data[data['Rainfall']>1]
raintod=data[data['RainToday'] == 'Yes']
raintod.equals(rain)

#dropping the equivalent colunms in order to decrese NaNs
data2=data.drop(columns=['Rainfall','RainToday','RISK_MM'])
data2.head()

#plotting Graphs to eliminate features that are not interdependent with RainTomoorrow
#barplot of Wind Gust Direction
plt.figure(figsize=(10,8))
sns.countplot(x=data2.WindGustDir,hue=data2.RainTomorrow)
plt.show()

#barplot of Wind Direction 9 am
plt.figure(figsize=(10,8))
sns.countplot(x=data2.WindDir9am,hue=data2.RainTomorrow)
plt.show()

#barplot of Wind Direction 3 pm
plt.figure(figsize=(10,8))
sns.countplot(x=data2.WindDir3pm,hue=data2.RainTomorrow)
plt.show()

#Eliminating the two not so dependent features
data3=data2.drop(columns=['WindDir9am','WindDir3pm'])
matrix=data3.corr()

#plotting a correlation matrix to see relation of variables 
plt.figure(figsize=(12,8))
sns.heatmap(data=matrix, annot=True)
plt.show()




```
